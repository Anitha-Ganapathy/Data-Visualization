```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
from pandas.plotting import scatter_matrix
%matplotlib inline
```

```python
# Import yfinance package
import yfinance as yf

# Set the start and end date
start_date = '1990-01-01'
end_date = '2022-05-20'

# Set the ticker
ticker = 'AMZN'
ticker_2 = 'GOOGL'
# Get the data
data = yf.download(ticker, start_date, end_date)
data_G = yf.download(ticker_2, start_date, end_date)
```

```python
start_date = '2012-11-1'
end_date = '2020-12-31'
data = data[data.index > start_date ]
data_G = data_G[data_G.index > start_date ]
```

## Perfect representation of data with context and labels for Amazon and Google stock prices. 

```python
# Plot adjusted close price data
fig , ax = plt.subplots(figsize=(12,7))
ax = data['Close'].plot(label = "AMZN")
ax = data_G['Close'].plot(label = "GOOGL")
plt.legend()
plt.title("Amazon and Google stock closing price over the years", fontsize = 20)
plt.show()
```

## Misrepresenting the data by not providing the title and the axis labels.
I have even omitted to provide the context by forgetting to mention that its for the Amazon and Google stock price beginning the year 2020 which is the start of Covid pandemic.

```python
fig , ax = plt.subplots(figsize=(12,5))
ax = data[data.index > '2020-01-01']['Close'].plot()
ax = data_G[data_G.index > '2020-01-01']['Close'].plot()
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
plt.show()
```

## Stock price fluctions and not mentioning the period for which it is represented

```python
fig , ax = plt.subplots(figsize=(8,5))
ax = data[data.index > '2021-01-01']['Close'].plot()
ax = data_G[data_G.index > '2021-01-01']['Close'].plot()
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
# Setting the interval of ticks of y-axis to 10.
listOf_Yticks = np.arange(0, 3800, 700)
plt.yticks(listOf_Yticks)
plt.show()
```

## Ways to mislead or misrepresent.
- Not providing the context
- Not displaying the axis ticks or simly put what the x and y axis labels are
